#### ML - Percentile :
Percentiles used to give you a `number` that `describes the value that a given percent of the values are lower than.` 

```python
#Let we have an array 
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31] 
```

In this case, What is the 75. percentile? The answer is 43, meaning that 75% of the people are 43 or younger.

```python
# NumPy percentile() method;
import numpy as np 
x = np.percentile(ages, 75)
x
```

```python
#Similarly, for the 90. percentile 
y = np.percentile(ages , 90)
y
```

#### ML - Data Distribution : 

```python
# lets taking an array containing 200 random floats between 0 and 10:
import numpy as np

x = np.random.uniform(0.0, 10.0, 200)
x
```

```python
import matplotlib.pyplot as plt

plt.hist(x,5)
```

The bars represents how many values in the array are between that interval(bar).
1. 44 values are between 0 and 2
2. 32 values are between 2 and 4
3. 40 values are between 4 and 6
4. 50 values are between 6 and 8
5. 33 values are between 8 and 10

##### Big Data distribution : 

```python
x = np.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 50)
plt.show()
```
