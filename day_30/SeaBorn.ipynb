#### Ploting using Seaborn 

```python
#Importing library
import seaborn as sns  
```

```python
#import data set
boat = sns.load_dataset('titanic')
boat.head()
```

```python
#to draw a bar plot
boat.survived.value_counts().plot(kind='bar')
```

```python
# OR
sns.countplot(x='survived', data=boat, hue='sex')
```

##### To plot a Scatter Plot:


```python
sns.scatterplot(x='age', y='fare', data=boat, hue='pclass')   
```

```python
sns.scatterplot(x='sex', y='fare', data=boat, hue='pclass')   
```

##### Ploting BarPlot: 

```python
sns.barplot(x='sex', y='fare', data=boat, hue='pclass')
```

##### Ploting PairPlot :

```python
sns.pairplot(boat)
```

##### Ploting Histogram :

```python
boat.hist() 
```

```python
boat.hist(bins=11, figsize=(10,7)) #figsize is used to increase the size of the graph and bins is used to increase the number of bars in the graph
```
