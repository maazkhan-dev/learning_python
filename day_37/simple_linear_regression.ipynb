## Machine Learning Algorithm
### 1. Simple Linear regression
#### Step 01: importing the Dataset


```python
# import dataset
import pandas as pd
df = pd.read_csv('Salary_Data.csv')
df.head()

```

#### Step 02: splitting the data into training & testing

```python
X = df[['YearsExperience']] #input will always in 2D
Y= df['Salary']
```

```python
X.head()
```

```python
Y.head()
```

```python
# importing Library for Model Training & splitting data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2,random_state=0)
```

### Step 03: Fit Linear regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train,Y_train)
model
```

#### Step 04: ploting

```python
import matplotlib.pyplot as plt

plt.scatter(X_train,Y_train)
plt.plot(X_train, model.predict(X_train), color='red')
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.title("Train plot")
plt.show()
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test, model.predict(X_test), color='orange')
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.title("Test plot")
plt.show()
```

#### Step 05: Testing Or Evaluating your Model 

```python
#Model Fitness
print("Score for training data :", model.score( X_train,Y_train))

```

```python
print("Score for testing data :", model.score( X_test,Y_test))
```
