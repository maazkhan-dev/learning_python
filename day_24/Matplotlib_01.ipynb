#### Learning Matplotlib (1)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

xpoints = np.array([0, 5])
ypoints = np.array([0, 25])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([2, 7, 0, 9])

plt.plot(ypoints, marker = '*') # Marker is used to specify the shape of the marker (*,o,+ etc.)
plt.show()
```

```python
# Dotted line
ypoints = np.array([3, 8, 1, 10])   

plt.plot(ypoints, marker = 'o', linestyle = 'dotted') # Linestyle is used to specify the type of the line (dashed, dotted, solid etc.)
plt.show()
```

##### Adding x-y Labels 

```python
x = np.array([2,3,4,5,6,7])
y = np.array([100000,200000,300000,400000,500000,600000])

plt.plot(x, y)

plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.show()
```

##### Adding title 

```python
x = np.array([2,3,4,5,6,7])
y = np.array([100000,200000,300000,400000,500000,600000])

plt.plot(x, y)

plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.show()
```
