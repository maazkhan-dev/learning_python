#### 13. splitting a string into multiple Columns

```python
import pandas as pd 
df = df = pd.DataFrame({"name": ['maaz khan','ammar ayaz','kashif haider','atif shah'],
                   "location": ['malakand, kpk','nowshehra, kpk','kohat, kpk ', 'karak, kpk']})
df
```

```python
#splitting a column into two columns
df['name'].str.split(' ', expand=True)
```

```python
#adding these split into new Columns
df[['first Name', '2nd name']] = df.name.str.split(' ',expand=True)
df
```

```python
df[['city','province']] = df.location.str.split(', ',expand=True)
df
```

```python
#refine data manupulation
df[['name', 'location', 'city', 'province']]
```

#### 14. Aggregate by multiple groups/function

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
#this shows no. of catagories 
df.groupby('sex').count()
df.groupby('who').count()
```

```python
len(df.who)
```

```python
len(df.groupby('who'))
```

```python
df.groupby(['sex','who','pclass']).count()
```

```python
# selecting specific columns
df[['sex','age','pclass']].head()
```

```python
df.describe()
```

```python
# for selecting specific no. of rows
df.describe().loc[['min','25%','50%', '75%', 'max']]
# Or 
df.describe().loc['min' : 'max']
```

```python
#to select specific rows and specific columns
df.describe().loc['min' : 'max', ['survived', 'age'] ]
df.describe().loc['min' : 'max', 'survived' : 'parch' ]
```
