## how to rename columns

```python
df=pd.DataFrame({'col A' : [1,2,3,4,5,6,7,8,],'col B' :[11,12,13,14,15,16,17,18] }) 
df.rename(columns={'col A' : 'Y','col B' : 'Z' }, inplace=True)  
df
```

```python
#OR easy renaming
df.columns=['Col__01', 'Col__02']
df
```

```python
#to replace character or string
df.columns=df.columns.str.replace('__', ' ')
df
```

```python
#To replace again 
df.columns=df.columns.str.replace(' ', '__')
df
```

```python
# Adding suffix to columns
df = df.add_suffix('_end')
df 
```

```python
# Adding prefix to columns
df = df.add_prefix('Baba__')
df 
```

```python
# back to original form after prefix and suffix
df.columns=['col_01','col_02']
df
```

### 16- Splitting a string into multiple columns

```python
import pandas as pd
df = pd.DataFrame( {'name' : ['maaz khan','ammar ayaz','kashif haider','atif shah' ],
            'location' : ['malakand, kpk','nowshehra, kpk','kohat, kpk ', 'karak, kpk'] } )
df
```

```python
#splitting a column into two columns
df.name.str.split(' ',expand=True)
```

```python
#adding these split into new Columns
df[['first Name', '2nd name']] = df.name.str.split(' ',expand=True)
df
```

```python
df[['city','province']] = df.location.str.split(', ',expand=True)
df
```

```python
#refine data manupulation
df[['name', 'location', 'city', 'province']]
```
