#### Learning Matplotlib (6) 

```python
#importing Libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([20,30,40,10])

thislabel = ['W','X','Y','Z']  
plt.pie(y, labels=thislabel)
plt.show()
```

##### Explode :

```python
thislabel = ['W','X','Y','Z']
thisExlpode = [0.2,0,0,0]  # explode 1st slice and 0.2 means 20% of the radius and so on

plt.pie(y, labels=thislabel, explode=thisExlpode)
plt.show()
```

##### Adding Shadow :


```python
plt.pie(y, labels=thislabel, explode=thisExlpode, shadow=True)  #   shadow = True will add shadow to the pie chart
plt.show() 
```

##### Adding Colours :

```python
mycolour = ['black','red','blue','Hotpink']  # we can also use hex code for colour like #FF5733

plt.pie(y, labels=thislabel, explode=thisExlpode, shadow=True, colors=mycolour)  
plt.show()
```

##### Adding legend :


```python
plt.pie(y, labels=thislabel, explode=thisExlpode)
plt.legend()
plt.show()
```

```python
# Legend with title/header
plt.pie(y, labels=thislabel, explode=thisExlpode)
plt.legend(title='four Different colours '    )
plt.show()
```
