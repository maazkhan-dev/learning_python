#### Learning Matplotlib (5)

##### Histograms :

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(100,10,50)
x
```

```python
plt.hist(x)
plt.show()
```

```python
#### Addiing bins to histogram
plt.hist(x, bins=20)
plt.show()  
```

##### Creating Pie Charts :


```python
y = np.array([30,45,35,45])
plt.pie(y)
plt.show()
```

##### Adding Labels :

```python
y = np.array([20,30,40,10])

thislabel = ['A','B','C','D']  
plt.pie(y, labels=thislabel)
plt.show()
```

##### Start Angles  :


```python
plt.pie(y, labels = thislabel, startangle = 270)
plt.show() 
```
