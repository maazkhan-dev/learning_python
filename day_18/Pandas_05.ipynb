#### 15. Reshape MultiIndex Series

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
df.survived.mean()
```

```python
df.groupby('sex').survived.mean()
```

```python
df.groupby(['sex', 'class']).survived.mean().unstack()  #unstack func just for separate class in column
```

#### 16. Contineous to catagorical data


```python
df.age.head()
```

```python
# Creating Bin (we are actually converting a range of age to a catagory) using pd.cut func
pd.cut(df.age, bins = [0,18,25,99], labels=['child', 'young', 'adult']). head()
df['new_age'] = pd.cut(df.age, bins = [0,18,25,99], labels=['child', 'young', 'adult']). head()
df.head()
```

#### 17. Converting One Set Of data into Another 

```python
df.sex.head()
```

```python
# we are using a dictionary map
df.sex.map({'male':0, 'female': 1})
df['sex_numbered'] = df.sex.map({'male':0, 'female': 1})
df.head()
```

```python
df.embarked.unique()
```

```python
df.embarked.factorize()[0]
df['embarked_numbered']= df.embarked.factorize()[0]
df.head()
```
