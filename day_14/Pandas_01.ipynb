##### Adding Column Names

```python
df=pd.DataFrame(np.random.rand(2, 4), columns=list('abcd'))
df
```

##### renaming Column Names

```python
df.rename(columns={'a':'A',  'b': 'B', 'c' : 'C', 'd' : 'D' }, inplace=True)
df
```

```python
#Same for DataFrames Created Through Pandas 
df=pd.DataFrame({'col A' : [1,2,3,4,5,6,7,8,],'col B' :[11,12,13,14,15,16,17,18] }) 
df.rename(columns={'col A' : 'Y','col B' : 'Z' }, inplace=True)  
df
```

```python
#OR easy renaming
df.columns=['Col__01', 'Col__02']
df
```

```python
#to replace character or string
df.columns=df.columns.str.replace('__', ' ')
df
```

##### Adding Prefix & Suffix

```python

# Adding suffix to columns
df = df.add_suffix('_end')
df 
```

```python
# Adding prefix to columns 
df = df.add_prefix('Start_')
df 
```
