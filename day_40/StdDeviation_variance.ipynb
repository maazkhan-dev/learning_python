#### Standard Deviation : 
S.D is a number that describes how spread out the values are.
OR : The square root of the variance; 
###### A low standard deviation means that most of the numbers are close to the mean (average) value.mA high standard deviation means that the values are spread out over a wider range.

```python
import numpy as np 
```

```python
#Using numpy std()
speed = [61,62,63,64,61,72,67,68]
x = np.std(speed)
x
```

```python
# Similarly if we Take a Large Range of number like 
speed = [32,111,138,28,59,77,97]
x = np.std(speed)
x
```

#### Variance :
Variance is another number that shows how spread out the data/values are.
If we multiply the S.D by itself, we get the variance ; i.e S.D^2



##### Manually Calculation :
To Calculate Variance we have to Follow Four Steps :
1. Find mean
2. For each value, find the difference from the mean
3. For each difference: find the square value
4. The variance is the average number of these squared differences:

```python
#1. Mean :
(61 + 62 + 63 + 64 + 61 + 72 + 67 + 68 ) / 8
```
