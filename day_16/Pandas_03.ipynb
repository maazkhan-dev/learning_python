#### 11. split dataFrames into two Subset

```python
import seaborn as sns
import pandas as pd 
df = sns.load_dataset('titanic')
df.head(3)
len(df)
```

```python

#we use From random import random 
from random import random
boat_01 = df.sample(frac=0.5, random_state=1)
boat_01.shape
```

```python
boat_02 = df.drop(boat_01.index)    #Droping boat_01 from the Overall df
boat_02.shape 
```

```python
boat_01.head()
```

```python
boat_02.head()
```

```python
len(boat_02) + len(boat_01)
```

#### 12. joining the Two DataSets

```python
df1 = pd.concat([boat_01, boat_02]) #this will joined the two kashties
df1.shape
```
