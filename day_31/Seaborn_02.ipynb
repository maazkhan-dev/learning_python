##### Ploting using Seaborn : 

##### Boxplot :

```python
import seaborn as sns
boat = sns.load_dataset('titanic')
```

```python
sns.boxplot(data=boat, x='sex', y='fare', hue='pclass')
```

```python
sns.boxenplot(data=boat, x='sex', y='fare', hue='pclass')
```

##### StripPlot :
###### Strip Plot is the Combination of Bar and Scattered Plot 

```python
sns.stripplot(data=boat, x='sex', y='fare', hue='pclass')
```

##### SwamPlot :

```python
sns.swarmplot(data=boat, x='sex', y='fare', hue='pclass')
```

##### Violin Plot : 

```python
sns.violinplot(data=boat, x='sex', y='fare', hue='pclass')
```
