#### 6. Reverse Row Order 

```python
import seaborn as sns
import pandas as pd
df= sns.load_dataset('titanic')
df.head()
```

```python
#it will reverse the sequence and will give you the head in reverse order
df.loc[::-1].head()
```

```python
#it will now just reset the index of df
df.loc[::-1].reset_index(drop=True).head()
```
