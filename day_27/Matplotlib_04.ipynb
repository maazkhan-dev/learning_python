#### Learning Matplotlib (4)

##### Bar Plots :

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array(["X", "Y", "Z"])
y = np.array([3, 8, 1])

plt.bar(x,y)
plt.show()
```

```python
x = ['book', 'mouse', 'keyboard']
y = [100,  300, 800]

plt.bar(x,y)
plt.show()

```

##### For Displaying bar Horizentally  :

```python
#Using horizontal bars barh() function
x = np.array(["X", "Y", "Z"])
y = np.array([3, 8, 1])

plt.barh(x,y)
plt.show()  
```

##### Adding Bar Colour :

```python
x = ['book', 'mouse', 'keyboard']
y = [100,  300, 800]

plt.bar(x,y, color = '#4CAF50') # can select color by name or by hex code
plt.show()  
```

##### Adjusting Bar Width :

```python
x = ['book', 'mouse', 'keyboard']
y = [100,  300, 800]

plt.bar(x,y, color = "#904CAF", width=0.2) # width of bar can be adjust b/w 0 to 1(scale)
plt.show()  
```

##### Adjusting Bar Hight : 

```python
x = ['book', 'mouse', 'keyboard']
y = [100,  300, 800]

plt.bar(x, y, color='#4CAF50', width=0.4) 

plt.title("Inventory Levels")
plt.ylabel("Quantity")
plt.show()
```
