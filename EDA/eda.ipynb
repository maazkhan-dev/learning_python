## 2. Bigger Picture of data

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum() #finding missing or null values
```

```python
df2 = df
```

```python
# Null values in percentage
df2.isnull().mean()* 100
# OR another formula for the %age of null values 
df2.isnull().sum() * 100 / len(df)
```

```python
## if there is more than 70% data is null it become meaningless for your data
# usually we remove this column
#sometimes this column is also important we can't remove it
```

```python
## how to deal with missing values 
## >> catagorical/object, numeric(float,int), bool
```

```python
df2['age'].unique()
```

```python
df2['age'].max()
```

```python
df2['age'].min()
```

```python
#in order to fill the missing space we have to fill it with the mean(average )of age
df2['age'].fillna(df2['age'].mean()) #if there is no "inplace_True" option". it will not filling this.


```

```python
df2.isnull().sum()
```

```python
df2["age"].fillna(df2["age"].mean(), inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
print(df2['age'].min())
print(df2['age'].max())
print(df2['age'].mean())
# thus, we replaced the missing values with the average/mean i.e 29.699117647
```

```python
df3 = df2
```

```python
df3.isnull().sum()
```

```python
#there are 75% + data in null in deck, so we remove this column
df3 = df3.drop(columns=['deck'])

```

```python
df3.head()
```

```python
df3.isnull().sum()
```

```python
df4 = df3
```

```python
df4['embark_town'].fillna(df4['embark_town'].mode()[0], inplace=True)
df4.isnull().sum()
```

```python
df5 = df4
```

```python
df5['embarked'].fillna(df5['embarked'].mode()[0], inplace=True)
```

```python
df5.isnull().sum()
```

```python
df5['age'].fillna(df5['age'].mean(), inplace=True)
```

```python
df5.isnull().sum()
```
