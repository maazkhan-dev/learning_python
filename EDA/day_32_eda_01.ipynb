##### 3. Dealing with Missing values : 

```python
df2= df
```

```python
df2['age'].unique()
```

```python
df2['age'].min()
```

```python
df2['age'].max()
```

```python
# Inordernto fill the missing Values, we have to fill it with the mean/average of that feature if it's numeric
df2['age'].fillna(df2['age'].mean()) #If there is no "inplace=true" it will not filling the null values
df2['age'].isnull().sum()
```

###### We have still the Same null values 

```python

df2['age'].fillna(df2['age'].mean(), inplace=True) 
```

```python
df2['age'].isnull().sum()
```

```python
df2.isnull().sum()
```

###### there is No Null/ 0 null values in age 

```python
print(df2['age'].min())
print(df2['age'].max())
print(df2['age'].mean())
# thus, we actually replaced the missing values with the average/mean i.e 29.699
```
